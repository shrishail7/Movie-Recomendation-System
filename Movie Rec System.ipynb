```python

import numpy as np

import pandas as pd
```

```python
movies = pd.read_csv('tmdb_5000_movies.csv')
credits = pd.read_csv('tmdb_5000_credits.csv')
```

```python
movies.shape
```

```python
credits.shape
```

```python
movies = movies.merge(credits,on='title')
movies.shape
```

```python
movies
```

```python
'''
genres
id
keywords 
overview
title 
cast 
crew
'''

movies = movies[['id','genres','keywords','overview','title','cast','crew']]
movies.head()
```

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

```python
movies.iloc[0].genres
```

```python
import ast

def convert(obj):
    L = []
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
movies['keywords'] = movies['keywords'].apply(convert)
```

```python

# taking only first three casts

def convert3(obj):
    L = []
    counter=0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python

def convertDirector(obj):
    L = []
    for i in ast.literal_eval(obj):
        if i['job']=='Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(convertDirector)
```

```python
movies.head()
```

```python
# now all data is preprocessed 
#create a new column named 'tag' in which all genres , keywords , overview , cast and crew

'''

overview is string convert it into list 

'''

movies['overview'] = movies['overview'].apply(lambda x:x.split())

```

```python
movies.head()
```

```python
movies['tag'] = movies['genres'] + movies['keywords'] + movies['overview'] + movies['cast'] + movies['crew'] 
```

```python
movies.head()
```

```python
'''

Remove all the space between words because
Sam Worthington   --> SamWorthington

to differentiate betwn 
[Sam Mendes] and [Sam Worthington]

'''
```

```python
movies['genres'] = movies['genres'].apply(lambda x : [i.replace(" ","") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x : [i.replace(" ","") for i in x])
movies['cast'] = movies['cast'].apply(lambda x : [i.replace(" ","") for i in x])
movies['crew'] = movies['crew'].apply(lambda x : [i.replace(" ","") for i in x])
```

```python
movies.head()
```

```python
movies = movies[['id','title','tag']]
```

```python
movies.head()
```

```python
#converting list into string 
movies['tag'] = movies['tag'].apply(lambda x:" ".join(x))
movies['tag']
```

```python
movies.head()
```

```python
#converting all into lower case
movies['tag'] = movies['tag'].apply(lambda x:x.lower())
```

```python
movies.head()
```

```python

'''
many of words like love , loved , loving are of same meaning so we have to replace all these types of words by base word

this is done by steamming nlti library

'''

from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()

def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)  #convert it into string again
```

```python
movies['tag'] = movies['tag'].apply(stem)
```

# converting text into vectors

```python
'''
By using method bag of words
other techniques
    1. word to vect
    2. if idf
    
    
How it work
 
 combine all the tags and then find frequency of each word in that combined tag take 5000 most frequent words
 now take each tag and find frequency of tag -> this is the vector of that movie
 
     action | science | fun | History ...(top 5000 words) -> most frequent words from combined tag 
 m1     4        2       1      0     ...  -> this is the vector of m1 movie
 m2     1        1       3      2     ...  -> this is the vector of m2 movie
 m3     0        1       1      4     ...  -> this is the vector of m3 movie
 .
 .
 .
 (5000 movies)
 (5000,5000) shape of vector
 
 
 IMPORTANT
 
 here we ignore the stop words 
 stop words -> used to frame the sentances but not related to meaning
 e.g. are,a,the,to,from
 
'''
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000 , stop_words='english')  #max_features max number of words , stop_words 
```

```python
vectors = cv.fit_transform(movies['tag']).toarray()     #converting transform to array
vectors
```

```python
vectors.shape
```

```python
# see the 5000 most frequent names
cv.get_feature_names()
```

```python
'''

Calculationg the distance between two vectors

here we are not interesed in uclidian distance , but in cosine distance
if cosine distance is small the two vectors are closer to each other , else enough long to each other

similarity ~ 1/distances

'''

from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vectors)   #similarity of movie with all others 
similarity                                #similarity itself is 1 else <1
```

```python
similarity.shape
```

```python
'''

steps to recomend the movie
1. a movie is given 
2. find index of that movie in our data in similarity
3. sort the similarity 
4. pick top 5 movies 

''' 
```

```python
#fetching the index
movies['title']=='Avatar'
movies[movies['title']=='Avatar']
movies[movies['title']=='Avatar'].index
movies[movies['title']=='Avatar'].index[0]
```

```python
#sorting without loosing the index  
list(enumerate(similarity[0]))
```

```python
sorted(list(enumerate(similarity[0])) ,reverse = True) #this sortting done on basis of index but we want on similarity
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])

# but we want first 5 movies

sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]

```

```python
def recomend(movi):
    movie_index = movies[movies['title']==movi].index[0]
    distances = similarity[movie_index]
    movies_list = sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:6]
    for i in movies_list:
        print(movies.iloc[i[0]].title)    
```

```python
recomend('Avatar')
```

```python
recomend('Batman Begins')
```
